```python
import networkx as nx
```

```python
graph = nx.DiGraph()
```

```python
graph.add_node('Node A')
graph.add_node('Node B')
graph.add_node('Node C')

graph.add_edge('Node A', 'Node B')

layout = nx.spectral_layout(graph)

nx.draw_networkx(graph, layout)
```

```python
nodes = 10
edge_prob = 0.2

ER = nx.erdos_renyi_graph(nodes, edge_prob, seed=1, directed=True)

layout = nx.circular_layout(ER)

nx.draw_networkx(ER, layout, arrowsize=20)
```

```python
ER.nodes(), ER.edges()
```

```python
dimension_1 = 4
dimension_2 = 4

graph = nx.hexagonal_lattice_graph(dimension_1, dimension_2)

layout = nx.spring_layout(graph)

nx.draw_networkx(graph, layout)
```

```python
matrix = nx.to_numpy_array(graph)

matrix.shape
```
